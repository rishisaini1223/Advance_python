```python
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
print(df.isnull().sum())
```

```python
#check percentange of attrition on depratment
# filter  the data to show only 'yes' values in the 'attrition 'column
```

```python
attr_df =df[df['Attrition']=='Yes']
```

```python
attr_df
```

```python
# calculction the attrition by department
```

```python
attrition_by_dpt = attr_df.groupby(['Department']).size().reset_index(name ="count")
attrition_by_dpt
```

```python
from plotly.graph_objs.layout import title
from plotly.graph_objs.pie import marker
# Create a donut chart
fig = go.Figure(data =[go.Pie(
    labels = attrition_by_dpt['Department'],
    values= attrition_by_dpt['count'],
    hole =0.4,
    marker = dict(colors =['red','green']),
    textposition = 'inside'
)])
# update the layout
fig.update_layout(title ='Atrrition by Department')
# show the chart
fig.show()
```

```python
attrition_by_fld = attr_df.groupby(['EducationField']).size().reset_index(name ="count")
attrition_by_fld
```

```python
fig = go.Figure(data =[go.Pie(
    labels = attrition_by_fld['EducationField'],
    values= attrition_by_fld['count'],
    hole =0.4,
    marker = dict(colors =['red','green','black']),
    textposition = 'inside'
)])

fig.update_layout(title ='Atrrition by EducationField')
fig.show()
```

```python
attrition_by = attr_df.groupby(['YearsAtCompany']).size().reset_index(name ='count')
attrition_by
```

```python
# we can see that most of the employees leave the organision
```

```python
attrition_by = attr_df.groupby(['YearsSinceLastPromotion']).size().reset_index(name ='count')
attrition_by
```

```python
attrition_by = attr_df.groupby(['Gender']).size().reset_index(name ='count')
attrition_by
# create a dount chart
fig = go.Figure(data=[go.Pie(
    labels = attrition_by['Gender'],
    values= attrition_by['count'],
    hole =0.4,
    marker = dict(colors =['red','green','black']),
    textposition = 'inside'
)])
# update the layout
fig.update_layout(title='Attrition by Gender')
# show the chart
fig.show()
```
