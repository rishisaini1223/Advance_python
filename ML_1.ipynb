```python
# Machine Learning  ---
# Data (50) --> x(Input data), y(target data)
# x ---> x_train , x_test
# y ---> y_train , y_test
# for machine training we will pass (x_train, y_train)
# for machine testing  we pass (x_test, y_test)

# 1+1 =2
# 1+2=3
# 1+3=4

# x = (1+1, 1+2, 1+3)
# y = (2,3,4)
# x_train = (1+1,1+2)
# x_test = (1+3)
# y_train = (2,3)
# y_test = (4)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/covid_toy - covid_toy.csv')
# df.head(3)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# fill null values
df['fever'] = df['fever'].fillna(df['fever'].mean())
```

```python
x = df.drop(columns = ['has_covid'])   # input data
# x
```

```python
y = df['has_covid'] # target or output data
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
print('total data shape: ',df.shape)
print('----------------------')
print('input data shape: ',x.shape)
print('x_train data shape: ',x_train.shape)
print('x_test data shape', x_test.shape)
print('-----------------------')
print('target data shape: ', y.shape)
print('y_tarin data shape: ',y_train.shape)
print('y_test data shape: ',y_test.shape)
```
