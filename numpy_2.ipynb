- hstack() and vstack()

- hstack() ----> it will add 2 or more arrays in harizontally.
- vstack() ----> it will add 2 or more arrays in vertically.

```python
a = np.arange(1,5).reshape(2,2)
b = np.arange(5,9).reshape(2,2)
c = np.arange(9,13).reshape(2,2)
```

```python
np.hstack((a,b,c))
```

```python
np.vstack((a,b,c))
```
